```python
#import dependencies 
import pandas as pd
import numpy as np
from sqlalchemy import create_engine
import psycopg2
```

```python
#load csvs


January = "January2018.csv"
February = "February2018.csv"
March = "March2018.csv"
April = "April2018.csv"
May = "May2018.csv"
June = "June2018.csv"
July = "July2018.csv"
August = "August2018.csv"
September = "September2018.csv"
October = "October2018.csv"
November = "November2018.csv"
December = "December2018.csv"
```

```python
#write to dataframes
January_df = pd.read_csv(January)
February_df = pd.read_csv(February)
March_df = pd.read_csv(March)
April_df = pd.read_csv(April)
May_df = pd.read_csv(May)
June_df = pd.read_csv(June)
July_df = pd.read_csv(July)
August_df = pd.read_csv(August)
September_df = pd.read_csv(September)
October_df = pd.read_csv(October)
November_df = pd.read_csv(November)
December_df = pd.read_csv(December)
```

```python
#bring in additional dataframes and csvs
neighborhoods = "neighborhood_codes.csv"
neighborhoods_df = pd.read_csv(neighborhoods)
buldings = "Building-Permits-by-Neighborhood-2018-2018-all.csv"
buildings_df = pd.read_csv(buldings)
```

```python
#view one month of data to get an idea of data
May_df.head()
```

```python
#concatinating dfs together into one master file. Through prelim data analysis , we concluded the data was well normalized 
master_df = pd.concat([January_df, February_df, March_df, April_df, May_df, June_df, July_df, August_df, September_df, October_df, November_df, December_df ], ignore_index=True)
```

```python
#view master df
master_df
```

```python
#drop unnecessary columns
clean1_df =  master_df.drop(columns=['FlagCrime', 'FlagUnfounded', 'FlagAdministrative', 'FlagCleanup', 'LocationName', 'LocationComment'])
```

```python
#only keep values >=1 because crimes that are -1 were later changed 
df_filtered = clean1_df[clean1_df['Count'] >= 1] 


```

```python
#view DF
df_filtered
```

```python
#write file to csv
df_filtered.to_csv("2018.csv")
```

```python
#line replaces blank space to NAN--If this is ``True`` then `to_replace` *must* be a string.
final_df = df_filtered.replace(r'^\s*$', np.nan, regex=True)
```

```python
#view df
final_df
```

```python
#view df
neighborhoods_df
```

```python
neighborhoods_mod=neighborhoods_df.rename(columns={"Neighborhood_Name":"nei_name"})
```

```python
neighborhoods_mod.head()
```

```python
#merge neighborhood table to masterfile 
df = pd.merge(final_df, neighborhoods_mod, how='right', on='Neighborhood')

df.head()
```

```python
#view df

mod_df=df.rename(columns={'Neighborhood_Name': 'nei_name'})
mod_df.head()
```

```python
#finding the unique values of the neighborhood column and lenght
names = mod_df.nei_name.unique()
len(names)
```
